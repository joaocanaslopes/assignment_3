##  Upload do ficheiro para o Google Colab

Nesta célula fazemos o upload do ficheiro CSV através da ferramenta `files.upload()`, que permite escolher um ficheiro guardado no computador.

Ao executar, aparece um botão para selecionar o ficheiro `kiva_loans.csv`.



```python
from google.colab import files
uploaded = files.upload()
```

##  Carregar o dataset com Pandas

Depois do upload, usamos `pd.read_csv()` para ler o ficheiro.

`list(uploaded.keys())[0]` devolve o nome do ficheiro enviado, permitindo carregá-lo automaticamente sem escrever o nome manualmente.

`df.head()` mostra as primeiras 5 linhas da tabela para confirmar que foi carregada corretamente.


```python
import pandas as pd

df = pd.read_csv(list(uploaded.keys())[0])
df.head()
```

# Question 1 – Select rows using string methods and numerical conditions


## 1.1 Filtrar linhas com empréstimos acima de 1000

Usamos uma condição numérica (`df['loan_amount'] > 1000`) para selecionar apenas os empréstimos cujo valor é superior a 1000.

`display()` serve para mostrar o resultado de forma organizada no Colab.


```python
display(df[df['loan_amount'] > 1000])
```

## 1.2 Selecionar empréstimos nos sectores Agriculture e Food

Aqui utilizamos o método `.isin()` para manter apenas as linhas em que o sector pertence à lista: *Agriculture* ou *Food*.

Isto é um exemplo de filtragem com métodos de strings.


```python
display(df[df["sector"].isin(["Agriculture","Food"])])
```

# Question 2 – Sort table using one or more columns


## 2.1 – Sort table using one or more columns

Nesta célula vamos ordenar o DataFrame pelo valor dos empréstimos (`loan_amount`) de forma crescente.  
Selecionamos apenas as colunas `country`, `sector` e `loan_amount` para mostrar os dados mais relevantes.  

O método usado é `.sort_values(by="loan_amount")`, que organiza as linhas do menor para o maior valor.


```python
display(df[["country", "sector", "loan_amount"]].sort_values(by="loan_amount"))
```

# Question 3 – Group and aggregate rows with `groupby`

## 3.1 Média do valor dos empréstimos por país

Aqui usamos `groupby("country")` para agrupar as linhas por país.

Depois aplicamos `.mean()` à coluna `loan_amount` para obter o valor médio de empréstimos em cada país.


```python
df.groupby("country")["loan_amount"].mean()
```

## 3.2 Soma total dos empréstimos por sector

Este código agrupa os dados pelo sector e calcula a soma total do valor dos empréstimos (`loan_amount`) dentro de cada sector.

É um exemplo de agregação com `groupby` e `.sum()`.


```python
df.groupby("sector")["loan_amount"].sum()
```

# Question 4 – Create a new variable with `apply` and `lambda`

```python
def categorize_loan(x):
    if x < 500:
        return "Small"
    elif x < 1500:
        return "Medium"
    else:
        return "Large"

df["loan_category"] = df["loan_amount"].apply(lambda x: categorize_loan(x))
```

```python
display(df[["loan_amount", "loan_category"]])
```
